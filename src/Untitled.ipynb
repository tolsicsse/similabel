```python
import pandas as pd
from glob import glob
import os
import numpy as np
```

```python
drivers = {}
for _ in glob('../data/*.csv'):
    file = os.path.split(_)[-1]
    file_split = file.split('_')
    driver = file_split[0]
    type = os.path.splitext(" ".join(file_split[2:]))[0]
    print(file, driver, type)
    if driver not in drivers:
        drivers[driver] = {}
    if type == 'events':
        events = pd.read_csv(_)
        events = events[['date_time','payload.type']]
        events['Lap'] = events['payload.type'].apply(lambda x: int(x.split("_")[0][-1]) if x.startswith('Lap') else 0)
        drivers[driver][type] = {'Lap':events['Lap'].values.tolist(), 'date_time':events['date_time'].values.tolist()}
    else:
        drivers[driver][type] = f"http://localhost:8081/{file}"
```

```python
data = []
for k,n in drivers.items():
    n['driver']=k
    data += [{"data":n}]
```

```python
with open("../drivers.json", "w") as fp:
    json.dump(data, fp)
```
